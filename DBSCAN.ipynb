# Neighborhood Approaches and Density Based Spacial Clustering with Applicated Noise (DBSCAN)

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X, y = make_blobs(n_samples= 1000, centers=12, n_features=2, 
                  random_state=333)
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
neighborhood_radius = [0.2, 0.7, 4, 8]
for ep in neighborhood_radius:
    db = DBSCAN(eps= ep, min_samples=10, 
                metric='euclidean')
    plt.title('Toy problem with eps: ' + str(ep))
    plt.scatter(X[:,0], X[:,1], c=db.fit_predict(X))
    plt.show()
```
